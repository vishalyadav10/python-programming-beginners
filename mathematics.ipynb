```python
#3_Check if a number is power of another number
def isPower (x, y):
    if (x == 1): 
        return (y == 1) 
    pow = 1
    while (pow < y):
         pow = pow * x 
    return (pow == y) 
t=int(input())
for _ in range(t):
    n,m=list(map(int,input().split()))
    if(isPower(n,m)): 
        print(1) 
    else: 
        print(0) 
  
```

```python
#ap_gp_fib
def ap(a):
    d=a[1]-a[0]
    for i in range(2,n):
        if (a[i]-a[i-1] != d):
            return False
    return True
def gp(a):
    m=a[1]//a[0]
    for i in range(2,n):
        if (a[i]/a[i-1] !=m):
            return False
    return True
def fib(a):
    flag = True
    for i in range(2,n):
        if a[i] == a[i-1]+a[i-2]:
            continue
        else:
            flag = False
            break
    return flag    
   
t=int(input())
for _ in range(t):
    n=int(input())
    a=[int(i) for i in input().split()[:n]]  
    if ap(a):
        d=a[1]-a[0]
        print(a[-1]+d)
    elif gp(a):
        m=a[1]/a[0]
        print(int(a[-1]*m))
    elif fib(a):
        print(a[-1]+a[-2])
    else:
        print("-9999")
    
        
```

```python
#trailing_zeroes_in_fact
t=int(input())
for _ in range(t):
    n=int(input())
    b=1
    while(n>1):
        b=b*n
        n-=1
    digit=[int(x) for x in str(b)]
    count=0
    digit.reverse()
    for i in range(len(digit)):
        if digit[i]==0:
            count+=1
        else:
            break
    print(count)       
```
